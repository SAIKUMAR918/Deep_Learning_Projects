```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def gen_data(n=500):
    x = np.random.randn(n, 2)
    y_ = np.product(x, axis=1) > 0
    y = np.array(y_, dtype=np.int8)
    return(x, y)
```

```python
x,y = gen_data()
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0], x[:,1], color=[['red', 'blue'][x]
                                               for x in y])
```

### Perceptron

```python
model = Sequential()
model.add(Dense(1, input_shape=(2,), activation='sigmoid'))
```

```python
# model.compile(optimizer='Adam',
#      loss='binary_crossentropy',
#      metrics=['accuracy'])

model.compile(optimizer='rmsprop',
      loss='binary_crossentropy',
      metrics=['accuracy'])
```

```python
# Training a perceptron model on the 2D data
model.fit(x, y, epochs=100)
```

```python
x_test=np.random.randn(1000,2)
y_ =np.product(x_test, axis=1)
y_test = np.where(y_<0,0,1)
```

```python
x_test
```

```python
preds = model.predict(x_test).reshape(-1,).astype('int8')
```

```python
pred1 = (model.predict(x_test) > 0.5).astype(int).reshape(-1)
```

```python
pred1
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.scatter(x_test[:,0], x_test[:,1], color=[['red', 'blue'][x] for x in pred1])
plt.title('Predictions')
plt.subplot(1,2,2)
plt.scatter(x_test[:,0], x_test[:,1], color=[['red', 'blue'][x] for x in y_test])
plt.title('Truth')
plt.show()
```

### MLP

```python
model1 = Sequential()
model1.add(Dense(12, input_shape=(2,), activation='tanh'))
model1.add(Dense(1, activation='sigmoid'))
```

```python
# model1.compile(optimizer='Adam',
#      loss='binary_crossentropy',
#      metrics=['accuracy'])

model1.compile(optimizer='rmsprop',
      loss='binary_crossentropy',
      metrics=['accuracy'])
```

```python
# Training the MLP on the 2D data
model1.fit(x, y, epochs=200)
```

```python
pred2 = (model1.predict(x_test) > 0.5).astype(int).reshape(-1)
```

```python
pred2
```

```python
# Ploting the predictions against the ground truth labels
plt.subplot(1,2,1)
plt.scatter(x_test[:,0], x_test[:,1], color=[['red', 'blue'][x] for x in pred2])
plt.title('Predictions')
plt.subplot(1,2,2)
plt.scatter(x_test[:,0], x_test[:,1], color=[['red', 'blue'][x] for x in y_test])
plt.title('Truth')
plt.show()
```
